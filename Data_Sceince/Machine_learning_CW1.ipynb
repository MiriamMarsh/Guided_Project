Objective: Creating a model to predict whether a given person lives in Yerushalayim or not based on certain features.

```python
import pandas as pd
```

```python
dira = pd.read_csv("Apartments Dataset .csv")
#data.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
print(dira['elevator(y/n)'].value_counts.isnull())
```

```python
dira= dira.drop('name', axis=1)
```

```python
dira.head()
```

```python
x = 'y'
if type(x) == str:
    print(1)
else:
    print(0)
```

```python
x= {'n': 0, 'y': 1}
dira["elevator(y/n)"] = dira["elevator(y/n)"].replace(x)
```

```python
x= {'n': 0, 'y': 1}
dira["makolet on your street?(y/n)"] = dira["makolet on your street?(y/n)"].replace(x)
```

```python
dira.head()
```

```python
dira['elevator(y/n)'] = dira['elevator(y/n)'].astype('bool')
```

```python
dira['makolet on your street?(y/n)'] = dira['makolet on your street?(y/n)'].astype('bool')
```

```python
x= {'n': 0, 'y': 1}
dira["Yerushalyim? (y/n)"] = dira["Yerushalyim? (y/n)"].replace(x)
```

```python
dira['Yerushalyim? (y/n)'] = dira['Yerushalyim? (y/n)'].astype('bool')
```

```python
dira.head(60)
```

Our target is weather the given person lives in Yerushaleim or not. Our features are the: Number of floors, wheather there is a elevator or not, number of apartments in the building and if there is a Makolet or not. 
In a markdown cell write one or two sentences with some analysis regarding the features and
targets.

```python
X= dira.drop('Yerushalyim? (y/n)', axis=1)
y=dira['Yerushalyim? (y/n)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.15)
```

```python
model = LogisticRegression(max_iter= 500, penalty='l2',C=.5)
model.fit(X_train, y_train)
pred= model.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(accuracy)
```

The accuracy of this model is 75%. We can try other other methods such as the KNN Neighbors and see if that would give us better results.
