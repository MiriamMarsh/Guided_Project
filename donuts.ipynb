```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import guided_project_functions as gpf
import matplotlib.style as style
```

```python
df = pd.read_csv("donuts.csv")
```

```python
style.use('classic')
df['type'].value_counts().plot.bar(color = 'purple', width = 0.3) 
#plt.bar(df['branch'], df["type"])
#(color = 'purple', width = 0.3)  #default width is 0.8
plt.title("Donut types sold")
plt.xlabel("Type of Donuts")
plt.ylabel("Number of Donuts Sold")
plt.xticks(rotation=20)
```

```python
# I would use the classic style for an audience which rather see which type sold the most at a quick glance.
```

```python
style.use('bmh')
df['type'].value_counts().plot.bar(color = 'red', width = 0.6) 
#plt.bar(df['branch'], df["type"])
#(color = 'purple', width = 0.3)  #default width is 0.8
plt.title("Donut types sold")
plt.xlabel("Type of Donuts")
plt.ylabel("Number of Donuts Sold")
plt.xticks(rotation=20)
```

```python
# The BMH style I would use for the owner to be able to see which types and exactly how many are being sold. The light grey boxes in the background are very helpful for a clearer analysis.
```
